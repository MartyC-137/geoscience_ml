```python
#import modules
import pandas as pd
import numpy as np
import plotly.express as px

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('analytes.csv')
df.head()
```

```python
df_ml = df[['MASTERID'] + [col for col in df if 'ICP' in col]]
print(df_ml.shape)
df_ml.head()
```

### Copper

```python
#copper
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
from sklearn.pipeline import Pipeline

X = df_ml.loc[:, ~df_ml.columns.isin(['Cu_ICP_PPM'])].drop('MASTERID', axis = 1)
y = df_ml['Cu_ICP_PPM']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import Ridge
import xgboost as xgb

print('Copper ML Scores: \n')

#random forest
rf_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', RandomForestRegressor(max_depth = 10,
                                    n_estimators = 200))
])
rf_pipeline.fit(X_train, y_train)
y_pred_rf = rf_pipeline.predict(X_test)

print('Random Forest: {:.3f}'.format(r2_score(y_test, y_pred_rf)))

#extra trees
et_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', ExtraTreesRegressor(max_depth = 10,
                                  n_estimators = 200))
])
et_pipeline.fit(X_train, y_train)
y_pred_et = et_pipeline.predict(X_test)

print('Extra Trees: {:.3f}'.format(r2_score(y_test, y_pred_et)))

#gradient boosting
gbr_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', GradientBoostingRegressor(max_depth = 10,
                                        n_estimators = 200))
])
gbr_pipeline.fit(X_train, y_train)
y_pred_gbr = gbr_pipeline.predict(X_test)

print('Gradient Boosting: {:.3f}'.format(r2_score(y_test, y_pred_gbr)))

#basic linear regression
lr = LinearRegression().fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print('Linear Regression: {:.3f}'.format(r2_score(y_test, y_pred_lr)))

#ada boost
ada = AdaBoostRegressor().fit(X_train, y_train)
y_pred_ada = ada.predict(X_test)

print('Ada Boost Score: {:.3f}'.format(r2_score(y_test, y_pred_ada)))

#bagging regressor
br = BaggingRegressor().fit(X_train, y_train)
y_pred_br = br.predict(X_test)

print('Bagging Regressor: {:.3f}'.format(r2_score(y_test, y_pred_br)))

#ridge
ridge = Ridge().fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

print('Ridge: {:.3f}'.format(r2_score(y_test, y_pred_ridge)))

#lasso
lasso = Lasso().fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)

print('Lasso: {:.3f}'.format(r2_score(y_test, y_pred_lasso)))

# #xgboost
# xgb = xgb.XGBRegressor(max_depth = 10).fit(X_train, y_train)
# y_pred_xgb = xgb.predict(X_test)

# print('XG Boost Score: {:.3f}'.format(r2_score(y_test, y_pred_xgb)))

xgb_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', xgb.XGBRegressor(max_depth = 10,
                               n_estimators = 200))
])
xgb_pipeline.fit(X_train, y_train)
y_pred_xgb = xgb_pipeline.predict(X_test)

print('XG Boost: {:.3f}'.format(r2_score(y_test, y_pred_xgb)))
```

```python
#lets blend the predictions of the 4 best performing models:
#this only works for regression models ------
def blended_model(x):
    return ((0.25 * rf_pipeline.predict(x)) + \
           0.25 * et_pipeline.predict(x) + \
           0.25 * gbr_pipeline.predict(x) + \
           0.25 * xgb_pipeline.predict(x))

y_pred_blended = blended_model(X_test)

print('Blended Model Score: {:.3f}'.format(r2_score(y_test, y_pred_blended)))
```

```python
#model accuracy = 72.6%
cu_df = pd.DataFrame({'Actual_Cu_PPM': y_test})
cu_df['Pred_Cu_PPM'] = y_pred_blended
cu_df.round(2).head()
```

```python
#add the MASTERID, latitude and longitude back in
cu_df = cu_df.merge(df[['MASTERID', 'LAT', 'LONG', 'STRAT']], left_index = True, right_index = True)
cu_df = cu_df[['MASTERID', 'LAT', 'LONG', 'STRAT', 'Actual_Cu_PPM', 'Pred_Cu_PPM']]
cu_df[['Actual_Cu_PPM', 'Pred_Cu_PPM']] = cu_df[['Actual_Cu_PPM', 'Pred_Cu_PPM']].astype('float').round(2)
print(cu_df.shape)
cu_df.head(20)
```

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://{user}:{pw}@localhost/{db}'.format(
                                                                           user = 'root',
                                                                           pw = 'martysql',
                                                                           db = 'quest_geochem_ml'))
```

```python
cu_df.to_sql('quest_copper_ml_blended',
         con = engine,
         if_exists = 'replace',
         index = False)
```

```python
cu_df.to_csv('quest_cu_ml_blended.csv')
```

### Zinc

```python
#zinc
X = df_ml.loc[:, ~df_ml.columns.isin(['Zn_ICP_PPM'])].drop('MASTERID', axis = 1)
y = df_ml['Zn_ICP_PPM']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print('Zinc ML Scores: \n')

#random forest
rf_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', RandomForestRegressor(max_depth = 10,
                                     min_samples_leaf = 3,
                                     min_samples_split = 4,
                                     n_estimators = 200))
])
rf_pipeline.fit(X_train, y_train)
y_pred_rf = rf_pipeline.predict(X_test)

print('Random Forest: {:.3f}'.format(r2_score(y_test, y_pred_rf)))

#extra trees
et_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', ExtraTreesRegressor(max_depth = 10,
                                     min_samples_leaf = 3,
                                     min_samples_split = 4,
                                     n_estimators = 200))
])
et_pipeline.fit(X_train, y_train)
y_pred_et = et_pipeline.predict(X_test)

print('Extra Trees: {:.3f}'.format(r2_score(y_test, y_pred_et)))

#gradient boosting
gbr_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', GradientBoostingRegressor(max_depth = 10,
                                     min_samples_leaf = 3,
                                     min_samples_split = 4,
                                     n_estimators = 200))
])
gbr_pipeline.fit(X_train, y_train)
y_pred_gbr = gbr_pipeline.predict(X_test)

print('Gradient Boosting: {:.3f}'.format(r2_score(y_test, y_pred_gbr)))

#basic linear regression
lr = LinearRegression().fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print('Linear Regression: {:.3f}'.format(r2_score(y_test, y_pred_lr)))

#ada boost
ada = AdaBoostRegressor().fit(X_train, y_train)
y_pred_ada = ada.predict(X_test)

print('Ada Boost Score: {:.3f}'.format(r2_score(y_test, y_pred_ada)))

#bagging regressor
br = BaggingRegressor().fit(X_train, y_train)
y_pred_br = br.predict(X_test)

print('Bagging Regressor: {:.3f}'.format(r2_score(y_test, y_pred_br)))

#ridge
ridge = Ridge().fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

print('Ridge: {:.3f}'.format(r2_score(y_test, y_pred_ridge)))

#lasso
lasso = Lasso().fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)

print('Lasso: {:.3f}'.format(r2_score(y_test, y_pred_lasso)))

# #xgboost
# xgb = xgb.XGBRegressor(max_depth = 10).fit(X_train, y_train)
# y_pred_xgb = xgb.predict(X_test)

# print('XG Boost Score: {:.3f}'.format(r2_score(y_test, y_pred_xgb)))

xgb_pipeline = Pipeline(steps = [
    ('data_scaling', StandardScaler()),
    ('model', xgb.XGBRegressor(max_depth = 10,
                               n_estimators = 200))
])
xgb_pipeline.fit(X_train, y_train)
y_pred_xgb = xgb_pipeline.predict(X_test)

print('XG Boost: {:.3f}'.format(r2_score(y_test, y_pred_xgb)))
```

```python
y_pred_blended = blended_model(X_test)

print('Blended Model Score: {:.3f}'.format(r2_score(y_test, y_pred_blended)))
```

```python
#model accuracy = 71.8%
zn_df = pd.DataFrame({'Actual_Zn_PPM': y_test})
zn_df['Pred_Zn_PPM'] = y_pred_blended
zn_df.round(2).head()
```

```python
#add the MASTERID, latitude and longitude back in
zn_df = zn_df.merge(df[['MASTERID', 'LAT', 'LONG', 'STRAT']], left_index = True, right_index = True)
zn_df = zn_df[['MASTERID', 'LAT', 'LONG', 'STRAT', 'Actual_Zn_PPM', 'Pred_Zn_PPM']]
zn_df[['Actual_Zn_PPM', 'Pred_Zn_PPM']] = zn_df[['Actual_Zn_PPM', 'Pred_Zn_PPM']].astype('float').round(2)
print(zn_df.shape)
zn_df.head(20)
```

```python
zn_df.to_sql('quest_zn_ml_blended',
         con = engine,
         if_exists = 'replace',
         index = False)
```
